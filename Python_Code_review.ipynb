```python
class Net(nn.Module):
  def __init__(self, imput_sz=1):
    super(Net, self).__init__()
    self.layer = nn.Linear(imput_sz, 1)
    self.input_sz = imput_sz

  def forward(self, x):
    x = self.layer(x)
    return x
```

```python
model1 = Net (imput_sz=10)
```

```python
model2 = Net()
```
